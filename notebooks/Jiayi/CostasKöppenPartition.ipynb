** KÖPPEN-GEIGER CLIMATE CLASSIFICATION **

In this notebook we present the partition of the stations based on the dataset that represents the Köppen-Geiger world map on climate classification. 

Bellow is a brief description we got from the website that we downloaded the dataset from:

The most frequently used climate classification map is that of Wladimir Köppen, presented in its latest version 1961 by Rudolf Geiger. A huge number of climate studies and subsequent publications adopted this or a former release of the Köppen-Geiger map. While the climate classification concept has been widely applied to a broad range of topics in climate and climate change research as well as in physical geography, hydrology, agriculture, biology and educational aspects, a well-documented update of the world climate classification map is still missing. Based on recent data sets from the Climatic Research Unit (CRU) of the University of East Anglia and the Global Precipitation Climatology Centre (GPCC) at the German Weather Service, we present here a new digital Köppen-Geiger world map on climate classification for the second half of the 20th century.

Koeppen-Geiger-ASCII.txt is the file that enables the classification of our base-stations. This dataset has a limited resolution, since it only uses two decimal places for each coordinate, so we had to round up the coordinates of our initial dataset accordingly. We expect that some of the base-station might be mistakenly assigned to a neighbouring area because of this limited precision.

```python
df2 = pd.read_csv("Koeppen-Geiger-ASCII.txt",  sep=r"\s+")
df2.columns = ['lat', 'long', 'Cls']
df2['lat'] = 4*df2.lat#, 1)
df2['long'] = 4*df2.long#, 1)
# df2[['latitude', 'longitude']] = df[['latitude', 'longitude']].astype(float)
df2.head()
```

Joining the two datasets...

```python
# pd.concat([df1, df2]).groupby(["key1", "key2"], as_index=False)
joined = pd.merge(df, df2, on=['lat','long'], how='inner')
shape(joined)
joined.head()

# a = joined.groupby(['Cls'])


# for index, row in a:
#     print row.index
```
