## **I.1 Climatic Classification**

A climatic classification may have a direct correlation to biology, as climatic conditions majorly influence life within a region. In literature, there are 2 famous weather classifications, **Köppen climate classification** and **Holdridge life zones system**.

## **I.2 Model and Method to study the climate classification**

We are using the PCA model on geographic parititions to study the climate classification. Based on the idea of Geographic Partitioning using KD tree, we are able to split the whole world into small parts with similar amount of measurements. However, the original algorithm has a major problem: some partitions are long and narrow shaped and therefore the cover areas that are expected to have completely different climatic conditions. We employed  two improvements to avoid this issue:



1. Before we utilized our kd-tree algorithm, we decided to split the globe in squares based on strictly geographical characteristics, thus grouping together regions that are expected to have somewhat similar weather conditions.
2. After forming these groups and before we employed our kd-tree algorithm we pruned areas that did not contain any measurements at all.

Moreover, when we were done with the steps described above, we performed kd-tree algorithm, to furtherly partition the square regions.

The following figure shows our result after performing these steps:

![Alt text](https://dl.dropboxusercontent.com/u/15585964/BigDataReportImages/allgroups.png "Partition")
