Then, let's do K means, initially we classify the data into 10 clusters. Again, each station can be considered as a point in 5D space, the 5 dimentions are:

1. Mean Temperature
2. Average difference of temperature
3. Mean Precipitation
4. Elevation
5. Latitude

```python
from sklearn.cluster import KMeans

def kmeans_climate(col,clusnum):
    table_Kmean=pd.read_pickle("kmeanclusterfull.pkl")
    mm=table_Kmean.as_matrix(columns=col)
    kmeans = KMeans(n_clusters=clusnum)
    kmeans.fit(mm)
    typelabel=kmeans.labels_.tolist()
    return typelabel
```

### 10 Clusters, using all 5D infos

The result looks good, it is pretty similar with the Köppen climate classification.

```python
from mpl_toolkits.basemap import Basemap
lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;

kmeans_climate(["mean_temp","mean_diff","mean_prcp","latitude","elevation"],10)
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='c')
m.drawcoastlines()
x,y = m(np.array(table_Kmean['longitude']), np.array(table_Kmean['latitude']))
color_map = m.scatter(x,y,c=np.array(typelabel),marker='.', lw=0, alpha=1)
m.colorbar(color_map,pad="10%")
```
