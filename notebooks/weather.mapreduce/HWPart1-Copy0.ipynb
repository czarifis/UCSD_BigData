#### We could be using mrjob.protocol to specify the type of the output file instead... Oh, well... Let's build a parser instead...

```python
'''
This is a parser that transforms our output to a csv file so that we can load it to pandas dataframes...
source: the source file (e.g counts)
dest: the destination file(e.g dest.csv)
'''
def generatecsv(source, dest):
    import csv
    myrow = []
    with open(source, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:

            for k in row:
                print k
                elements=k.split('\t')
                basestation= elements[0]
                maxmin= elements[1]
                #print maxmin
                maxmin = maxmin.replace('"', '')
                print maxmin
                maxslashmin=maxmin.split('~')
                maxcount = maxslashmin[0]
                mincount = maxslashmin[1]

            myrow.append([basestation, maxcount, mincount])


    with open(dest, 'w') as fp:
        a = csv.writer(fp, delimiter=',')
        a.writerows(myrow)
```

```python
generatecsv('countsEMR','basestationmaxmin.csv')
```
