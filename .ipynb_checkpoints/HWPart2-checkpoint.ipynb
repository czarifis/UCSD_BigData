####Apparently there are base-stations that have no TMAX-TMIN measurements at all, since the number of base station in the two dataframes vary. We chose to do an inner join in order to prune these base-stations

```python
res = pd.merge(df, maxmindf, on='station', how='inner')
res.head()
```

```python
#tempdf = res.groupby(pd.cut(res["latitude"], np.arange(-90, 90, 10)))#.head()#.sum()  #np.arange(0, 1.0+0.155, 0.155)
tempdf = res.copy()
tempdf['latgroup'] = pd.cut(res["latitude"], np.arange(-90, 90, 100))#.head()#.sum()  #np.arange(0, 1.0+0.155, 0.155)
tempdf['longroup'] = pd.cut(res["longitude"], np.arange(-180, 180, 100))#.head()#.sum()  #np.arange(0, 1.0+0.155, 0.155)
#tempdf.reset_index()

#tempdf['status'] = pd.cut(df['Days Late'], bins=[-1, 14, 35, 56, 365])


tempdf.dropna() 
tempdf.head()
```

```python
# http://matplotlib.org/basemap/users/merc.html

from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt


lons=tempdf.ix[:,'longitude'].values
lats=tempdf.ix[:,'latitude'].values
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# lat_ts is the latitude of true scale.
# resolution = 'i' means use intermediate resolution coastlines.
lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')

# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
# labels = [left,right,top,bottom]
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

#m.drawparallels(np.arange(-90.,91.,30.))
#m.drawmeridians(np.arange(-180.,181.,60.))
m.drawmapboundary(fill_color='aqua')

# draw map with markers for locations
x, y = m(lons,lats)
m.plot(x,y,'g.')

# cx1 = -10
# cy1 = -10
# cx2 = 10
# cy2 = 10

for i in range(len(x1z)):
    cx1 = x1z[i]
    cy1 = y1z[i]
    cx2 = x2z[i]
    cy2 = y2z[i]



    x1,y1 = map(cx1,cy1) 
    x2,y2 = map(cx1,cy2) 
    x3,y3 = map(cx2,cy2) 
    x4,y4 = map(cx2,cy1) 
    p = Polygon([(x1,y1),(x2,y2),(x3,y3),(x4,y4)],
        facecolor='red',edgecolor='white',linewidth=2) 
    plt.gca().add_patch(p) 

plt.title('weather stations')
plt.show()
```
